```python
## importing Libraries

import torch
import torch.nn as nn
import torch.optim as optim
from torchvision import datasets, transforms, models
from torch.utils.data import DataLoader, random_split
import matplotlib.pyplot as plt
import numpy as np
import time
```

```python
# Transforming data to 224*224 pixels
transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
])
```

```python
#Importing the dataset , Training size 80% and validation set 20%
train_dataset = datasets.CIFAR10(root='./data', train=True, download=True, transform=transform)
train_size = int(0.8 * len(train_dataset))
val_size = len(train_dataset) - train_size
train_data, val_data = random_split(train_dataset, [train_size, val_size])

test_data = datasets.CIFAR10(root='./data', train=False, download=True, transform=transform)

train_loader = DataLoader(train_data, batch_size=32, shuffle=True)
val_loader = DataLoader(val_data, batch_size=32, shuffle=False)
test_loader = DataLoader(test_data, batch_size=32, shuffle=False)
```

```python
# pretrained models implementation
models_dict = {
    "ResNet-18": models.resnet18(pretrained=True),
    "AlexNet": models.alexnet(pretrained=True),
    "MobileNetV2": models.mobilenet_v2(pretrained=True),
    "SqueezeNet": models.squeezenet1_1(pretrained=True),
    "VGG-16": models.vgg16(pretrained=True)
}
```

```python
# defining a loop with all the model names
for model_name, model in models_dict.items():
    if model_name == "ResNet-18":
        model.fc = nn.Linear(model.fc.in_features, 10)
    elif model_name == "MobileNetV2":
        model.classifier[1] = nn.Linear(model.classifier[1].in_features, 10)
    elif model_name == "AlexNet" or model_name == "VGG-16":
        model.classifier[6] = nn.Linear(model.classifier[6].in_features, 10)
    elif model_name == "SqueezeNet":
        model.classifier[1] = nn.Conv2d(512, 10, kernel_size=(1, 1), stride=(1, 1))
        model.num_classes = 10
```

```python

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
```

```python
# defining a function and evaluation all the models
def train_and_evaluate(model, model_name):
    model = model.to(device)
    criterion = nn.CrossEntropyLoss()
    optimizer = optim.Adam(model.parameters(), lr=0.001)

    train_losses = []
    val_accuracies = []

    # Training loop
    for epoch in range(5):  # Using a small number of epochs for demonstration
        model.train()
        running_loss = 0.0
        for images, labels in train_loader:
            images, labels = images.to(device), labels.to(device)

            optimizer.zero_grad()
            outputs = model(images)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()

            running_loss += loss.item()

        avg_loss = running_loss / len(train_loader)
        train_losses.append(avg_loss)
        print(f"Epoch [{epoch+1}/5], Loss: {avg_loss:.4f}")

        # Evaluation
        model.eval()
        correct = 0
        total = 0
        with torch.no_grad():
            for images, labels in test_loader:
                images, labels = images.to(device), labels.to(device)
                outputs = model(images)
                _, predicted = torch.max(outputs.data, 1)
                total += labels.size(0)
                correct += (predicted == labels).sum().item()

        accuracy = 100 * correct / total
        val_accuracies.append(accuracy)
        print(f"Validation Accuracy of {model_name}: {accuracy:.2f}%")

    # Plot training loss and validation accuracy
    plt.figure(figsize=(12, 5))
    plt.subplot(1, 2, 1)
    plt.plot(train_losses, label="Training Loss")
    plt.title(f"Training Loss for {model_name}")
    plt.xlabel("Epoch")
    plt.ylabel("Loss")
    plt.legend()

    plt.subplot(1, 2, 2)
    plt.plot(val_accuracies, label="Validation Accuracy")
    plt.title(f"Validation Accuracy for {model_name}")
    plt.xlabel("Epoch")
    plt.ylabel("Accuracy (%)")
    plt.legend()
    plt.show()

    # Return final evaluation metrics for this model
    return train_losses[-1], val_accuracies[-1]


evaluation_results = {}

for model_name, model in models_dict.items():
    print(f"\nTraining and evaluating {model_name}...")
    final_loss, final_accuracy = train_and_evaluate(model, model_name)
    evaluation_results[model_name] = {
        "Final Training Loss": final_loss,
        "Final Validation Accuracy": final_accuracy
    }

print("\n--- Model Evaluation Summary ---")
for model_name, metrics in evaluation_results.items():
    print(f"\nModel: {model_name}")
    print(f"  Final Training Loss: {metrics['Final Training Loss']:.4f}")
    print(f"  Final Validation Accuracy: {metrics['Final Validation Accuracy']:.2f}%")
```
